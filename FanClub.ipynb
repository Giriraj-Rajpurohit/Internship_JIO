```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import pymongo
import datetime
```

```python
df = pd.read_csv("cast.Shows(Aggeragated).csv")
```

```python
fanclub_page_urls = []

# Function to get cast page URL
def get_fanclub_page_url(show_url):
    try:
        response = requests.get(show_url)
        response.raise_for_status()  # Raise an error for bad status codes
        soup = BeautifulSoup(response.content, 'html.parser')
        fanclub_link = soup.find('a', {'title': 'FANCLUB'})
        if fanclub_link:
            return 'https://www.indiaforums.com' + fanclub_link['href']
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL {show_url}: {e}")
    return None

# Iterate over each show URL and get the cast page URL
for index, row in df.iterrows():
    show_url = row['Show URL']
    fanclub_page_url = get_fanclub_page_url(show_url)
    fanclub_page_urls.append(fanclub_page_url)
```

```python
if len(fanclub_page_urls) < len(df):
    fanclub_page_urls.extend([None] * (len(df) - len(fanclub_page_urls)))
```

```python
import requests
from bs4 import BeautifulSoup

def get_usernames(base_url):
    all_usernames = []
    page_number = 1

    while True:
        
        url = f"{base_url}?pn={page_number}" if page_number > 1 else base_url

        print(f"Fetching URL: {url}")

        
        response = requests.get(url)

        
        if response.status_code != 200:
            print(f"Failed to retrieve page {page_number}")
            break

        
        soup = BeautifulSoup(response.content, 'html.parser')

        
        user_elements = soup.find_all('p', class_='celeb-fanclub__name js-user')

        
        usernames = [user_element.text.strip() for user_element in user_elements]

        print(f"Usernames found on page {page_number}: {usernames}")

        
        if not usernames:
            print("No usernames found, stopping.")
            break

        
        all_usernames.extend(usernames)

        page_number += 1

    return all_usernames
```

```python

for index, row in df.iterrows():
    fanclub_url = row['Fanclub Page URL']
    if fanclub_url:  
        usernames = fetch_usernames(fanclub_url)
        usernames_str = ', '.join(usernames)  # Join usernames into a single string
        df.at[index, 'Fans'] = usernames_str  # Assign the string to the 'Fans' column
    else:
        df.at[index, 'Fans'] = ""  # Assign an empty string if the URL is None
```

```python

client = pymongo.MongoClient("mongodb://localhost:27017")
df_new = df.loc[:, ['_id', 'Show Name', 'Show URL', 'Fanclub Page URL', 'Fans']]
data = df_new.to_dict(orient="records")
db = client["Fans"]  

def insert_in_chunks(data, chunk_size=1000):
    for i in range(0, len(data), chunk_size):
        chunk = data[i:i + chunk_size]
        try:
            db.Shows.insert_many(chunk, ordered=False)
        except Exception as e:
            print(f"Error inserting chunk starting at index {i}: {e}")


insert_in_chunks(data)

print("Data insertion completed.")
```
